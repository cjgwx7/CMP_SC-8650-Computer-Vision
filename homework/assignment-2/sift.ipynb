# Implementation of **S**cale **I**nvariant **F**eature **T**ransform (SIFT) in Python

## Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Read Images

```python
img = cv2.imread("images/assignment3-2b.png", cv2.IMREAD_COLOR)
plt.imshow(img)
```

The default option for OpenCV is to read images in as BGR. Thus, the image was converted to RGB.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## SIFT Example

```python
### Initialize a SIFT object
sift = cv2.SIFT_create()

### Detect key points
kp = sift.detect(img, None)

### Draw keypoints on image
img_out = cv2.drawKeypoints(img, kp, img)

### Display SIFT image
plt.imshow(img_out)
```

```python
### Compute descriptors
kp, des = sift.detectAndCompute(img, None)
print(des.shape)
```
